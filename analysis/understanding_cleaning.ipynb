**Крок 1** завантаження бд

```python
import pandas as pd

campaigns = pd.read_csv("../Data/campaigns.csv")
customers = pd.read_csv("../Data/customers.csv")
events = pd.read_csv("../Data/events.csv")
products = pd.read_csv("../Data/products.csv")
transactions = pd.read_csv("../Data/transactions.csv")
```

**Крок 2** загальний аналіз 

```python
tables = {
    "campaigns": campaigns,
    "customers": customers,
    "events": events,
    "products": products,
    "transactions": transactions
}

for name, df in tables.items():
    print(f"\n{'='*20} {name.upper()} {'='*20}")
    df.info(show_counts=True)
```

**Крок 3** перші рядки

```python
tables = {
    "campaigns": campaigns,
    "customers": customers,
    "events": events,
    "products": products,
    "transactions": transactions
}

for name, df in tables.items():
    print(f"\n{'='*20} {name.upper()} {'='*20}")
    print(df.head())
```

**Крок 4** тип даних

```python
tables = {
    "campaigns": campaigns,
    "customers": customers,
    "events": events,
    "products": products,
    "transactions": transactions
}

for name, df in tables.items():
    print(f"\n{'='*20} {name.upper()} {'='*20}")
    print(df.dtypes)
```

**Крок 5** пропуски

```python
tables = {
    "campaigns": campaigns,
    "customers": customers,
    "events": events,
    "products": products,
    "transactions": transactions
}

for name, df in tables.items():
    print(f"\n{'='*20} {name.upper()} {'='*20}")
    print(df.isna().sum())
```

**Крок 6** дублікати

```python
tables = {
    "campaigns": campaigns,
    "customers": customers,
    "events": events,
    "products": products,
    "transactions": transactions
}

for name, df in tables.items():
    print(f"\n{'='*20} {name.upper()} {'='*20}")
    print(df.duplicated().sum())
```

**Крок 7** унікальність primary key

```python
print(f"campaigns - campaign_id: {campaigns["campaign_id"].duplicated().sum()}")
print(f"customers - customer_id: {customers["customer_id"].duplicated().sum()}")
print(f"events - event_id: {events["event_id"].duplicated().sum()}")
print(f"products - product_id: {products["product_id"].duplicated().sum()}")
print(f"transactions - transaction_id: {transactions["transaction_id"].duplicated().sum()}")
```

**Крок 8** перевірка категорій

```python
print(f"\n{'='*20} {"customers".upper()} {'='*20}\n")
print(f"{customers["gender"].value_counts()}\n")
print(f"{customers["country"].value_counts()}\n")
print(f"{customers["loyalty_tier"].value_counts()}\n")
print(f"\n{'='*20} {"events".upper()} {'='*20}\n")
print(f"{events["event_type"].value_counts()}\n")
print(f"{events["device_type"].value_counts()}\n")
print(f"{events["traffic_source"].value_counts()}\n")
print(f"\n{'='*20} {"products".upper()} {'='*20}\n")
print(f"{products["category"].value_counts()}\n")
print(f"{products["is_premium"].value_counts()}\n")
print(f"{products["base_price"].value_counts().sort_index(ascending=False)}\n")
print(f"\n{'='*20} {"transactions".upper()} {'='*20}\n")
print(f"{transactions["quantity"].value_counts()}\n")
print(f"{transactions["discount_applied"].value_counts()}\n")
print(f"{transactions["gross_revenue"].value_counts()}\n")
print(f"{transactions["refund_flag"].value_counts()}")
```
